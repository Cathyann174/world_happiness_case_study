**Rename columns**
As a data cleaning step rename the columns as needed. Standardize the column names, so that they are all snake_case, correct column names that are misspelt, and make column names more concise as needed.

```python
# Rename columns in df2015

df2015.rename(columns={'Country':'country'}, inplace=True)
df2015.rename(columns={'Happiness Rank':'happiness_rank'}, inplace=True)
df2015.rename(columns={'Happiness Score':'happiness_score'}, inplace=True)
df2015.rename(columns={'Economy (GDP per Capita)':'economy'}, inplace=True)
df2015.rename(columns={'Family':'family'}, inplace=True)
df2015.rename(columns={'Health (Life Expectancy)':'health'}, inplace=True)
df2015.rename(columns={'Freedom':'freedom'}, inplace=True)
df2015.rename(columns={'Trust (Government Corruption)':'trust'}, inplace=True)
df2015.rename(columns={'Generosity':'generosity'}, inplace=True)

# Display updated column names
print(df2015.columns)
```

```python
# Rename columns in df2016
df2016.rename(columns={'Country':'country'}, inplace=True)
df2016.rename(columns={'Happiness Rank':'happiness_rank'}, inplace=True)
df2016.rename(columns={'Happiness Score':'happiness_score'}, inplace=True)
df2016.rename(columns={'Economy (GDP per Capita)':'economy'}, inplace=True)
df2016.rename(columns={'Family':'family'}, inplace=True)
df2016.rename(columns={'Health (Life Expectancy)':'health'}, inplace=True)
df2016.rename(columns={'Freedom':'freedom'}, inplace=True)
df2016.rename(columns={'Trust (Government Corruption)':'trust'}, inplace=True)
df2016.rename(columns={'Generosity':'generosity'}, inplace=True)

print(df2016.columns)
```
